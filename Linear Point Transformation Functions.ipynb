# Import Libraries

```python
from skimage import data
from scipy.misc import imread, imresize
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

# Linear Point Transformation

Formula of Linear Point Transformation is:
s=r*c+b

```python
c=2
b=32
```

# Original Image

```python
original_image =imread('sydney.png',True, 'L')    #read image as grey scale image
```

# Processed Image

```python
processed_image =original_image.copy()
```

```python
#s=c*r+b
processed_image=c*processed_image +b
```

```python
#Normalizing Intensity Levels
processed_image[processed_image<0]=0
processed_image[processed_image>256]=256
```

```python
fig, axes = plt.subplots(1, 2)
ax = axes.ravel()

ax[0].imshow(original_image,cmap=plt.cm.gray,interpolation='bilinear')
ax[0].set_title("Original")
ax[1].imshow(processed_image, cmap=plt.cm.gray,interpolation='bilinear')
ax[1].set_title("pixel operation")
plt.imshow(processed_image,cmap=plt.cm.gray) 
```
